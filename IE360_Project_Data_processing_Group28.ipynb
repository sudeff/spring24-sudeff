## Preparing Data

In this file, we prepared data, converted into wide format, and tried some basic approaches. None of them were useful. Other approaches and the final approach is coded in R in a different file.

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm
from sklearn.linear_model import Lasso
```

```python
production = pd.read_csv('production.csv')
print(production["date"].dtype, production["hour"].dtype)
production
```

```python
production["date"] = pd.to_datetime(production["date"])
print(production["date"].dtype)
```

```python
print(len(production["date"].unique()))
```

```python
weather_info_long = pd.read_csv('weather_info.csv')
print(weather_info_long["date"].dtype, weather_info_long["hour"].dtype)
weather_info_long
```

```python
weather_info_long['date'] = pd.to_datetime(weather_info_long['date'])
print(weather_info_long["date"].dtype)
print(len(weather_info_long["date"].unique()))
weather_info_long.sort_values(by=['date', 'hour'])
```

### Converting weather_info_long into wide format

```python
melted = weather_info_long.melt(id_vars=['date', 'hour', 'lat', 'lon'], var_name='variable', value_name='value')
melted
```

```python
melted['variable_lat_lon'] = melted['variable'] + "_" + melted['lat'].astype(str) + "_" + melted['lon'].astype(str)
weather_info = melted.pivot_table(index=['date', 'hour'], columns='variable_lat_lon', values='value', aggfunc='first')
weather_info.reset_index(inplace=True)
weather_info
```

```python
data = pd.merge(weather_info, production, on=['date', 'hour'], how='outer')
data = data.sort_values(by=['date', 'hour'])
data = data.reset_index(drop=True)
data
```

```python
data['hour'] = pd.to_timedelta(data['hour'], unit='h')
data['datetime'] = data['date'] + data['hour']
data['hour'] = data['datetime'].dt.hour.astype(int)
data.insert(0, 'datetime', data.pop('datetime'))
data
```

```python
n_of_days = len(data["date"].unique())
n_of_days
```

```python
n_of_datapoints = n_of_days*24 - 8
n_of_datapoints
```

We have an excessive 48 rows in the data. Let's look for duplicated values.

```python
# Group the data by 'date' and count the number of unique values in 'hour' column for each group
hour_counts_per_date = data.groupby('date')['hour'].nunique()

# Find the dates with more than 24 unique hours
dates_with_extra_hours = hour_counts_per_date[hour_counts_per_date > 24]

# Print the dates with more than 24 unique hours and the corresponding count of unique hours
print("Dates with more than 24 unique hours:")
print(dates_with_extra_hours)
```

```python
# Check for duplicated rows in the DataFrame
duplicated_rows = data[data.duplicated()]

# Print the duplicated rows, if any
if not duplicated_rows.empty:
    print("Duplicated rows:")
    print(duplicated_rows)
else:
    print("No duplicated rows found.")
```

```python
data = data.drop_duplicates()
data = data.reset_index(drop=True)
data
```

## Investigating Data

### Finding the number of days, when the production amount is zero for each hour

```python
data.dropna(inplace=True)

zero_production_rows = data[data['production'] == 0]

zero_production_count_per_hour = zero_production_rows.groupby('hour').size()

zero_production_count_per_hour
```

We will discard the hours, whose production mean is less than 5% of the overall production mean. Discarded hours will be predicted as zero in the forecasting phase.

```python
prod_mean = data["production"].mean()
prod_mean_wout_zeros = data[(data['production'] != 0)]["production"].mean()
print(prod_mean_wout_zeros)

lower_limit = prod_mean_wout_zeros*5/100 
print(lower_limit)
```

```python
for i in range(24):
    mean = data[data["hour"] == i]["production"].mean()
    print(i, "Mean of hour is ", mean)
```

Now, we will discard the hours whose production will be predicted as zero from the model. These hours are: 0, 1, 2, 3, 4, 5, 18, 19, 20, 21, 22, 23.

```python
hours_to_discard = [0, 1, 2, 3, 4, 5, 18, 19, 20, 21, 22, 23]

filtered_data = data[~data['hour'].isin(hours_to_discard)]
filtered_data = filtered_data.reset_index(drop=True)

filtered_data
```

```python
filtered_data.dropna(inplace=True)

filtered_data.to_csv('filtered_data.csv', index=False)

filtered_data
```

## Regression with all variables

```python
feature_columns = [col for col in filtered_data.columns if col not in ['datetime', 'date', 'production']]

X = filtered_data[feature_columns]
y = filtered_data['production']

# Add a constant term to the predictor variable (intercept)
X = sm.add_constant(X)

# Fit the linear regression model
model = sm.OLS(y, X).fit()

# Print the regression summary
print(model.summary())
```

Summary information for CSNOW_surface variables:

```python
# Extract variable names that start with "CSNOW_SURFACE"
csnow_surface_variables = [var for var in model.params.index if var.startswith("CSNOW_surface")]

# Print summary information for each CSNOW_SURFACE variable
for var_name in csnow_surface_variables:
    print(f"Variable: {var_name}")
    print(f"Coefficient: {model.params[var_name]}")
    print(f"Standard Error: {model.bse[var_name]}")
    print(f"t-value: {model.tvalues[var_name]}")
    print(f"p-value: {model.pvalues[var_name]}")
    print("\n")
```

```python
coefficients = [model.params[var] for var in csnow_surface_variables]
pvalues = [model.pvalues[var] for var in csnow_surface_variables]

# Create a DataFrame to store coefficients and p-values
csnow_df = pd.DataFrame({'Variable': csnow_surface_variables, 'Coefficient': coefficients, 'p-value': pvalues})

# Calculate weights based on significance (inverse of p-values)
csnow_df['Weight'] = 1 / csnow_df['p-value']

# Normalize weights so that they sum up to 1
csnow_df['Weight'] = csnow_df['Weight'] / csnow_df['Weight'].sum()

#csnow_df = csnow_df.sort_values(by='p-value', ascending=True)

csnow_df
```

```python
CSNOW_weighted = []

for index, row in filtered_data[csnow_surface_variables].iterrows():

    # Calculate the weighted average for the current row
    weighted_average = sum(row[col] * weight for col, weight in zip(row.index, csnow_df["Weight"]))
    CSNOW_weighted.append(weighted_average)

filtered_data['CSNOW_weighted'] = CSNOW_weighted
```
